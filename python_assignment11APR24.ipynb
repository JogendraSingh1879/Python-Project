```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title SkinThickness vs Insulin

from matplotlib import pyplot as plt
dib.plot(kind='scatter', x='SkinThickness', y='Insulin', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Pregnancies vs Glucose

from matplotlib import pyplot as plt
dib.plot(kind='scatter', x='Pregnancies', y='Glucose', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Glucose vs BloodPressure

from matplotlib import pyplot as plt
dib.plot(kind='scatter', x='Glucose', y='BloodPressure', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(True)
```

```python
# @title Pregnancies

from matplotlib import pyplot as plt
dib['Pregnancies'].plot(kind='hist', bins=20, title='Pregnancies')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='BMI', y='Glucose', data=dib)
plt.ylabel('Glucose ($)')

plt.show()
```

```python
pd.crosstab(dib['Glucose'],dib['BloodPressure'])
```

```python
sns.stripplot(x='BMI', y='Age', data=dib, size=4,jitter=False)
plt.ylabel('Age ($)')
plt.show()
```

```python
sns.stripplot(x='BMI', y='Age', data=dib, size=4,jitter=True)
plt.ylabel('Age ($)')
plt.show()
```
